# Insertion Sort Algorithm

```javascript
let insertionSort = (array) => {

    for (let i = 1; i < array.length; i++) {

          /*
          This is the element which be moved. This create element with not data so the elements to the left 
          can be copied to this index if needed. The elements to the left will be moved until you find new 
          location for this element in the array.  
          */

          //Move the Current Element in tempory storage
          let currentElement = array[i]; 
          let lastIndex = i - 1;


          //Shifts the elements to the right. 
          while (lastIndex >= 0 && array[lastIndex] > currentElement) {
            array[lastIndex + 1] = array[lastIndex];
            lastIndex--;

          }

         //Found laction to put Current Element   
         array[lastIndex + 1] = currentElement;
    }
    
     return array;
}
```

```javascript
let arr = new Array(6,5,3,1,8,7,2,4);

let result = insertionSort( arr);


console.log( result.join() )
```

# Print Array

```javascript
function printerArray(arrayList){
    console.log(`_________________________________`);
    for( var i = 0; i < arrayList.length;i++){
        console.log(`[ ${i} ] = ${arrayList[i]}`);
    }
}

printerArray(new Array(6,5,3,1,8,7,2,4));
```

# Insertion Sort

```javascript
arr = new Array(6,5,'MOVING VALUE = 3',1,8,7,2,4); 
```

We are going to move 'MOVING VALUE = 3' at index 2. This is the outer *for loop* that you using to check all the value to left. So compare 6 and 5 to the 'MOVING VALUE = 3'. 

```javascript
i = 2;
```

```javascript
arr[i]
```

### Store Current Value in Temporary Variable.

Now we need the store the 'MOVING VALUE = 3' becase going move all values to the left to to right.

```javascript
currentElement = arr[i];
```

### Shift Elements to the Right

```javascript
lastIndex = i - 1;
```

```javascript
console.log( arr.join() );
```

```javascript
arr[lastIndex + 1] = arr[lastIndex];
```

Now we copyied [ 1 ] = 5 to the index 2. So [1] = 5 and [2] = 5 have the **same values**.

```javascript
console.log( arr.join() );
```

### Copy Current Value in the array

```javascript
arr[lastIndex ] = currentElement;
```

```javascript
console.log( arr.join() );
```
